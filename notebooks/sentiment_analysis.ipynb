## Returns Analysis

**Sector Adjusted Returns (i.e. "Excess" or "Active" Returns)**
Every return number is measured in excess of a given company's sector return. This (loosely) accounts for the market effects and the sector/style effect.

**Evaluate Trailing Returns & Forward Returns**

For each conference call, collect the returns of the company leading up to the call, and track the returns following the call

`[t-6 months, t-3 months, t-1 month]`... **`[call date at time t]`** ... `[t+6 months, t+3 months, t+1 month]`

**Quintile Return Spreads**
_**How do the (past and future) return profiles of top rated calls differ from the return profiles of bottom rated calls?**_

1. `Q1 Sentiment Returns`: Take the top 20% of calls and calculate the average prior returns (across that subset of high rated companies)
2. `Q5 Sentiment Returns`: Take the bottom 20% of calls and calcluate the average prior returns (across that subest of low rated companies).
3. `Quintile Spread`: Q1 - Q5

Note: This is not a real implemental strategy at this point; just intended for a quick analysis to study sentiment correlations/relationships with returns.

### Top Sentiment vs Bottom Sentiment: Quintile Spreads

```python
ret_windows = [21, 63, 127]
ret_col_pri = [f't-{rw} trading days' for rw in reversed(ret_windows)]
ret_col_fwd = [f't+{rw} trading days' for rw in ret_windows]
ret_cols = ret_col_pri + ret_col_fwd

ntiles = 5
call_data['Loughran & Mcdonald Q-Sprd'] = pd.qcut(call_data['Loughran & Mcdonald'], ntiles, labels=False)
call_data['TextBlob Q-Sprd'] = pd.qcut(call_data['TextBlob'], ntiles, labels=False)
call_data['FinBERT Q-Sprd'] = pd.qcut(call_data['FinBERT'], ntiles, labels=False)
```

```python
call_data.head()
```

```python
output = call_data.loc[call_data['call_date'] <= '2020-10-01',
                     ['cid', 'Loughran & Mcdonald Q-Sprd', 'TextBlob Q-Sprd', 'FinBERT Q-Sprd'] + ret_cols] \
    .melt(id_vars=['cid'] + ret_cols,var_name='measure',value_name='qtile')

output = (output.loc[output['qtile'].isin([0,4]),:]
          .groupby(['measure','qtile'])[ret_cols]
          .mean()*100).reset_index()\
            .melt(id_vars=['measure','qtile'])\
            .pivot(index=['measure','variable'],columns='qtile',values='value').reset_index()

output['quintile spread'] = output[4] - output[0]
```

### Trailing Returns of Sentiment Score Quintiles 
Forward Return of Top 20% of Calls minus Return of Bottom 20% of calls

```python
output_pri = output.pivot(index='variable',columns='measure',values='quintile spread').loc[ret_col_pri,:]
output_pri.plot(kind='bar',fontsize=15,rot=0)
plt.xlabel('Period Before & After Call', fontsize=20)
plt.ylabel('Excess Returns over Sector', fontsize=20)
plt.legend(prop={'size':20})
```

### Forward Returns of Sentiment Score Quintiles 
Forward Return of Top 20% of Calls minus Return of Bottom 20% of calls

```python
output_fwd = output.pivot(index='variable',columns='measure',values='quintile spread').loc[ret_col_fwd,:]
output_fwd.plot(kind='bar',fontsize=15,rot=0)
plt.xlabel('Period Before & After Call', fontsize=20)
plt.ylabel('Excess Returns over Sector', fontsize=20)
plt.legend(prop={'size':20})
```
