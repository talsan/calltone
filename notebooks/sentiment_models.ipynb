## Sentiment Model #1 (The OG): The Loughran & Mcdonald Financial Dictionary

### Background

### What is Dictionary-Based (aka "Bag-of-Words") Sentiment Analysis?

Count occurances of positive or negative words in a given text. A collection of words is called a "dictionary."

1. `Positive Sentiment`: Count all the positive words within a given text (e.g. a single conference call). 
2. `Negative Sentiment`:   Count all the negative words within a given text.
3. `Net Sentiment` = `Positive Sentiment` - `Negative Sentiment`

_see demo below_

### Pros & Cons of Dictionary-Based Methods

**Pros**
- Great for introductory tutorials like this one
- Easy to implement/understand
- Computationally cheap

**Cons**
- Over simplifies reality and completely lacks context (like what words surrounded the given word)
- List construction is subjective and lacks clear ways of measuring effectiveness

### Tokenization
**Definition:** Split longer strings of text into smaller pieces (tokens). Tokens can be individual words, multiple words, or sentences, and so on.

**Example:** "Hi there. My name is Tal." --> ["Hi", "there", "my", "name", "is", "Tal"]

**Code Below:** Tokenize and remove puncutation

````python
call_text_lm = call_text.copy()
call_text_lm['text'] = call_text_lm['text'].apply(lambda x: [w for w in tokenize(x, deacc=True)])
call_text_lm['text'][0][0:10]
````

### Stop-Words & Cleanup
**Definition:** Stop-words are words that contribute little to the overall meaning of a given text and often removed in the pre-processing step.

**Example:** "the", "and", "of", etc. 

**Code Below**: Remove Stop Words, Upper-Case Words, and Abnormally Short/Long Words

````python
call_text_lm['text'] = call_text_lm['text'].apply(lambda x: [remove_stopwords(w) for w in x])
call_text_lm['text'] = call_text_lm['text'].apply(lambda x: [w for w in x
                                               if (2 < len(w) < 15) and (w.islower())])
call_text_lm['text'][0][0:10]
````

### Stemming
**Definition:** Process of removing suffixes, prefixes, etc. from a word in order to obtain it's stem. Generally not as effective as "lemmatization" which more accurately captures a words meaning.

**Example:** "running" + "runner" + "runs" --> "run"

**Code Below:** Porter Stemmer on Document Text

````python
# stemming (porter)
p = PorterStemmer()
call_text_lm['text'] = call_text_lm['text'].apply(lambda x: p.stem_documents(x))
call_text_lm['text'][0][0:10]
````

Bag of Words Transformation

````python
# create dictionary object
dictionary = corpora.Dictionary(call_text_lm['text'])

# filter extremes
# dictionary.filter_extremes(no_below=2, no_above=0.5)

# bag-of-words transformation
corpus = [dictionary.doc2bow(text) for text in call_text_lm['text']]
````

### Load and Stem Dictionaries

````python
pos_list = pd.read_csv('./model_inputs/lm_word_lists/lm_positive.csv', header=None)[0].str.lower().to_list()
neg_list = pd.read_csv('./model_inputs/lm_word_lists/lm_negative.csv', header=None)[0].str.lower().to_list()
pos_neg = pos_list + neg_list

pos_stemmed = list(set(p.stem_documents(pos_list)))
neg_stemmed = list(set(p.stem_documents(neg_list)))
````

````python
pos_list[0:20]
````

````python
neg_list[0:20]
````

### Loughran & Mcdonald Equal-Weighted Method

Words are equall weighted (e.g. every pos/neg word carries the same importance)

**`Positive Sentiment`**: Positive words in a call / total words in call

**`Negative Sentiment`**: Negative words in a call / total words in call

**`Overall Sentiment`**: Positive Sentiment - Negative Sentiment

````python
word_cf = [[(dictionary[id], count) for id, count in line] for line in corpus]

# calculate count frequency sentiment
sent_cf = []
for call in word_cf:
    word_count = sum([cf for tkn, cf in call])
    pos = 0; neg = 0
    
    if word_count > 0:
        pos = sum([cf for tkn, cf in call if tkn in pos_stemmed]) / word_count
        neg = sum([cf for tkn, cf in call if tkn in neg_stemmed]) / word_count

    sent_cf.append({'word_count':word_count,
                    'pos_cf': pos,
                    'neg_cf': neg,
                    'sent_cf': pos - neg})
````

### Loughran & Mcdonald TF-IDF Method (For Completeness)

**TF-IDF**: Words that appear across many documents (document frequency) are downweighted.

This is recommended (over raw counts) in nearly all use-cases, *but is it right for this context?*

````python
# tfidf transformation
tfidf = models.TfidfModel(corpus)  # fit model
corpus_tfidf = tfidf[corpus]  # apply model

# count words per call
word_tfidf = [[(dictionary[id], count) for id, count in line] for line in corpus_tfidf]

# calculate tfidf sentiment
sent_tfidf = []
for call in word_tfidf:
    pos = sum([tfidf for tkn, tfidf in call if tkn in pos_stemmed])
    neg = sum([tfidf for tkn, tfidf in call if tkn in neg_stemmed])
    
    safe_div0 = (pos + neg) > 0

    sent_tfidf.append({'pos_tfidf': (pos / (pos + neg)) if safe_div0 else 0,
                       'neg_tfidf': (neg / (pos + neg)) if safe_div0 else 0,
                       'sent_tfidf': ((pos - neg) / (pos + neg)) if safe_div0 else 0})
````

````python
def calc_call_sentiment(predictions):
    sent_count = len(predictions)
    pos = predictions[predictions == 'positive'].count()
    neg = predictions[predictions == 'negative'].count()
    return pd.Series({'finBert_pos': pos / sent_count,
                      'finBert_neg': neg / sent_count,
                      'finBert_sent': (pos - neg) / sent_count
                      })


# take mean from sentences into one score per call
calls_sent_finBert = sent_finBert.groupby(['cid']).apply(lambda x: calc_call_sentiment(x['prediction']))
calls_sent_finBert['finBert_sent_mean'] = sent_finBert.groupby(['cid'])['sentiment_score'].mean()
````
